```python
import pandas as pd

df = pd.read_csv('./grades.csv')

df['平均成绩'] = df.iloc[:,2:].mean(axis=1)
df['等第成绩'] = pd.qcut(df['平均成绩'], [0, 0.25, 0.5, 0.75, 1], labels=['D', 'C', 'B', 'A'])
df = df.sort_values(by=['平均成绩'], ascending=False)
result_df = df[['学号', '姓名', '等第成绩']]
result_df.to_csv('./graded.csv', index=True)
result_df
```

```python
import chardet
# 检测文件编码
with open('email_text.txt', 'rb') as file:
    raw_data = file.read()
    result = chardet.detect(raw_data)
    encoding = result['encoding']
    print(encoding)
```

```python
import re

email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

with open('email_text.txt', 'r',encoding="GB2312") as file:
    content = file.read()

emails = re.findall(email_pattern, content)

protected_emails = [email.replace(email.split('@')[1], '[protected]', 1) for email in emails]

with open('extracted_emails.txt', 'w', encoding='utf-8') as file:
    for email in protected_emails:
        file.write(email + '\n')
```
